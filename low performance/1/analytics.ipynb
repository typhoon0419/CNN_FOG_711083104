```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
data_path = os.getcwd() + "/dataset/total_concentrated.csv"
features = ["A_F", "A_V", "A_L", "L_F", "L_T", "L_L", "T_F", "T_V", "T_L"]

dataset = pd.read_csv(data_path)
dataframe = dataset.loc[:, "A_F":"T_L"]
# dataset = dataset[["A_F", "A_V", "A_L", "L_F", "L_V", "L_L", "T_F", "T_V", "T_L", "Action"]].values

window_length = int(1*64)
total_windows = int((len(dataset))/window_length)
```

```python
dataframe = (dataframe-dataframe.mean())/dataframe.std()
```

```python
x = dataset.loc[window_length:window_length*2-1,"time"]
y = dataset.loc[window_length:window_length*2-1,"A_V"]

y1 = dataset.loc[window_length:window_length*2-1,"L_V"]
y2 = dataset.loc[window_length:window_length*2-1,"T_V"]
```

```python
plt.style.use("ggplot") 

plt.plot(x, y, 'r')
plt.plot(x, y1, 'g-')
plt.plot(x, y2, 'b-')
```
